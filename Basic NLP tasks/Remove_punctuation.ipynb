```python
import string

text = "Hello! How are you? I am learning NLP, it's interesting."

# Remove punctuation
clean_text = text.translate(str.maketrans('', '', string.punctuation))

print(clean_text)
```

**Remove Punctuation from Tokenized Words (Common in NLP)**

```python
import string
import nltk
from nltk.tokenize import word_tokenize

# Download the 'punkt_tab' resource if not already present
nltk.download('punkt_tab')

text = "Hello! How are you? I love NLP."

words = word_tokenize(text)

clean_words = [word for word in words if word not in string.punctuation]

print(clean_words)
```
